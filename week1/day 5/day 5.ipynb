```python
from pyspark.sql.types import *

schema = StructType([
    StructField("emp_id", IntegerType(), True),
    StructField("name", StringType(), True),
    StructField("salary", IntegerType(), True),
    StructField("address", StringType(), True),
    StructField("loc", StringType(), True),
    StructField("email", StringType(), True)
])
```

```python
df = spark.createDataFrame(data, schema)
df.show()
df.printSchema()
```

```python
df.select("emp_id","name","salary").display()
```

```python
from pyspark.sql.functions import *
df.select(col("emp_id"),col("name"),col("salary")).display()
```

```python
df1 = df.withColumn("bonus", col("salary") * 0.10)
df1.show()
```

```python
df2 = df.withColumn("salary", col("salary") * 1.20)
df2.show()
```

```python
df3 = df.withColumn(
    "salary_category",
    when(col("salary") >= 7000, "High")
    .when(col("salary") >= 6000, "Medium")
    .otherwise("Low")
)
df3.show()
```

```python
data1 = [
    (101, "Alice", 5000),
    (102, "Bob", 6000)
]

data2 = [
    (103, "Charlie", 5500),
    (104, "David", 7000)
]

cols = ["emp_id", "name", "salary"]

df1 = spark.createDataFrame(data1, cols)
df2 = spark.createDataFrame(data2, cols)

df_union = df1.union(df2)
df_union.show()
```

```python
df_upper = df.withColumn("name_upper", upper(col("name")))
df_upper.show()
```
